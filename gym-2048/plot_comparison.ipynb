```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
baseline_scores = np.loadtxt(r".\logs\baseline.csv",delimiter=",", dtype=int)
print(baseline_scores)
```

```python
monte_carlo = np.loadtxt(r".\logs\monte_carlo.csv",delimiter=",", dtype=int)
print(monte_carlo)
```

```python
x1 = np.linspace(0,len(baseline_scores),len(baseline_scores), dtype=int)
print(x1)
x2 = np.linspace(0,len(monte_carlo), len(monte_carlo), dtype=int)
print(x2)
plt.plot(x2, monte_carlo, label='monte_carlo')
plt.plot(x1, baseline_scores, label='baseline')

plt.legend()
plt.xlabel('Number of moves')
plt.ylabel('Score')
plt.show()
```

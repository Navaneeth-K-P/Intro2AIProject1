```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
monte_5 = np.loadtxt(r".\logs\monte_carlo_5.csv",delimiter=",", dtype=int)
monte_10 = np.loadtxt(r".\logs\monte_carlo_10.csv",delimiter=",", dtype=int)
monte_15 = np.loadtxt(r".\logs\monte_carlo_15.csv",delimiter=",", dtype=int)
monte_30 = np.loadtxt(r".\logs\monte_carlo_30.csv",delimiter=",", dtype=int)
monte_50 = np.loadtxt(r".\logs\monte_carlo_50.csv",delimiter=",", dtype=int)
```

```python
def seperate_sims(a):
    sims = []
    dum = []
    for i in range(0,len(a)-1):
        if(a[i]<=a[i+1]):
            dum.append(a[i])
        else:
            # print("lol")
            dum.append(a[i])
            sims.append(dum)
            dum = []
    dum.append(a[len(a)-1])
    sims.append(dum)
    return(sims)
```

```python
monte_5 = seperate_sims(monte_5)
monte_10 = seperate_sims(monte_10)
monte_15 = seperate_sims(monte_15)
monte_30 = seperate_sims(monte_30)
monte_50 = seperate_sims(monte_50)
```

```python
def avg_generator(a):
    min_len = min(len(a[0]), len(a[1]), len(a[2]), len(a[3]))
    a_dum = [a[0][:min_len], a[1][:min_len], a[2][:min_len], a[3][:min_len]]
    a_dum = np.asarray(a_dum)
    return (a_dum[0] + a_dum[1] + a_dum[2] + a_dum[3])/4
```

```python
def max_sims(a):
    return(max(len(a[0]), len(a[1]), len(a[2]), len(a[3])))
```

```python
def max_avg_generator(a):
    res = np.zeros((max_sims(a),), dtype=int)
    for i in range(0,len(res)):
        m = 0
        c = 0
        for j in range(0,4):
            if len(a[j]) > i:
                m += a[j][i]
                c += 1
        res[i] = m/c

    return res
```

```python
m5 = avg_generator(monte_5)
aug5 = max_avg_generator(monte_5)
# print(len(m5[1]))
print(max_sims(monte_5))
print(len(aug5))

m15 = avg_generator(monte_15)
aug15 = max_avg_generator(monte_15)
# print(len(m15[1]))
print(max_sims(monte_15))

m10 = avg_generator(monte_10)
aug10 = max_avg_generator(monte_10)
# print(len(m10[1]))
print(max_sims(monte_10))

m30 = avg_generator(monte_30)
aug30 = max_avg_generator(monte_30)
# print(len(m30[1]))
print(max_sims(monte_30))

m50 = avg_generator(monte_50)
aug50 = max_avg_generator(monte_50)
# print(len(m50[1]))
print(max_sims(monte_50))
```

```python
x5 = np.linspace(0,len(m5),len(m5), dtype=int)
x10 = np.linspace(0,len(m10),len(m10), dtype=int)
x15 = np.linspace(0,len(m15),len(m15), dtype=int)
x30 = np.linspace(0,len(m30),len(m30), dtype=int)
x50 = np.linspace(0,len(m50),len(m50), dtype=int)




plt.plot(x30, m30, label='max_sims = 30')
plt.plot(x50, m50, label='max_sims = 50')
plt.plot(x15, m15, label='max_sims = 15')
plt.plot(x10, m10, label='max_sims = 10')
plt.plot(x5, m5, label='max_sims = 5')

plt.legend()
plt.xlabel('Number of moves')
plt.ylabel('Score')
plt.show()
```
